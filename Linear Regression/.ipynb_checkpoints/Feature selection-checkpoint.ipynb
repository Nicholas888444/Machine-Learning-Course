# How to select features/variables in linear regression?
In linear regression, selecting the right variables (features) to include in your model is a crucial step. The goal is to choose the subset of variables that have the most significant impact on the dependent variable while avoiding multicollinearity (high correlation between independent variables). Next we will introduce two ways of selecting independent variables: correlation analysis and p-value.

### Correlation Analysis:
Calculate the correlation coefficients between all independent variables and the dependent variable. Variables with higher absolute correlation coefficients (positive or negative) are generally more important.

Seaborn's heatmap function can be used to visualize the correlation between variables in a DataFrame.

```python
# install seaborn
!pip install seaborn
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
# Load and Prepare Data
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Boston.csv')
df.head()
```

```python
# clean the data by dropping missing values (nan)
df = df.dropna()
```

#### Calculate Correlation Matrix:
Use the .corr() method on your DataFrame to calculate the correlation matrix, which contains correlation coefficients between pairs of variables.

```python
correlation_matrix = df.corr()
```

#### Create a Heatmap:
Use the sns.heatmap() function to create the heatmap. Pass the correlation matrix to the function, and customize the appearance of the heatmap as needed.

```python
plt.figure(figsize=(10, 8))  # Set the figure size (optional)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

 - **annot=True** adds the correlation values to the cells.  
 - **cmap** specifies the color map for the heatmap. You can choose from various color maps.  
 - **fmt=".2f"** formats the displayed values with two decimal places.  

#### Interpreting Correlation Values

Interpreting correlation values involves understanding the strength and direction of the relationship between two variables. Correlation coefficients range from -1 to 1, where,  

Positive Correlation: When the correlation coefficient is close to 1, it indicates a strong positive correlation. This means that as one variable increases, the other tends to increase as well. For example, if the correlation coefficient is 0.8, it suggests a strong positive relationship.  

Negative Correlation: When the correlation coefficient is close to -1, it indicates a strong negative correlation. This means that as one variable increases, the other tends to decrease. For example, if t e correlation coefficient is -0.6, it suggests a strong negative relationshi  .

No Correlation: When the correlation coefficient is close to 0, it suggests little to no linear relationship between the variables. Even if there is a relationship, it may be nonlinear. For example, if the correla ion coefficient is 0.1, it suggests a weak or negligible linear relations  hip.

From the above correlation heatmap, we can select several variables whose correlation value to the dependent variable (MEDV) is greater than 0.5 or less than -0.5, which indicates a strong positive or negative correlation.

The variables selected are as follows:

Variable    Correlation  
**RM** (average number of rooms per dwelling): 0.7    
**PTRATIO** (pupil-teacher ratio by town): -0.51  
**LSTAT** (lower status of the population): -0.74  

Next, we can build and fit a linear regression model based on the selected variables.

```python
Y = df['MEDV']
X = df[['RM', 'PTRATIO', 'LSTAT']]
# Add a constant term to the independent variables matrix for the intercept
X = sm.add_constant(X)

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 

# Fit the multiple linear regression model
model = sm.OLS(Y_train, X_train).fit()

summary = model.summary()
print(summary)
```

Next, we evaluate the model on the testing dataset:

```python
Y_pred = model.predict(X_test) # Generate predictions

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r_squared = r2_score(Y_test, Y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
print(f'R-squared (R): {r_squared:.2f}')
```

From the result, we can see the R-squared on the training/testing set is 0.69/0.63. This indicates a little overfitting.

Next, let's compare with a full model with all independent variables.

```python
Y = df['MEDV']
X = df.drop(['MEDV'],axis=1)
# Add a constant term to the independent variables matrix for the intercept
X = sm.add_constant(X)

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 

# Fit the multiple linear regression model
model = sm.OLS(Y_train, X_train).fit()
#model = sm.OLS(Y, X).fit()

summary = model.summary()
print(summary)

Y_pred = model.predict(X_test) # Generate predictions

mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r_squared = r2_score(Y_test, Y_pred)

print('\nEvaluation on Testing Set:')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
print(f'R-squared (R): {r_squared:.2f}')
```

From the result, we can see R-squared on the training/testing set is 0.75/0.67. This indicates a significant overfitting.  
The r2 of the simplier model is 0.63, which is close to 0.67. This indicates that by selecting only three independent variables that are mostly correlated to the dependent variable, we can get a simpler model of the similar prediction accuracy but with less overfitting and easier to interpret.

### P-Value
We can also select independent variables by p-value. In linear regression, the p-value associated with each independent variable provides a measure of the statistical significance of that variable's contribution to explaining the variance in the dependent variable. Typically, if the p-value is less than your chosen significance level (e.g., 0.05), you would conclude that there is a statistically significant relationship between the independent variable and the dependent variable. Conversely, if the p-value is greater than the significance level, you would fail to make the conclusion.

We can adopt one of the following variable/feature selection methods using p-value:  
**Forward Selection**: Start with an empty model and iteratively add variables based on their contribution to model performance.  
**Backward Elimination**: Start with all variables and iteratively remove the least significant ones.  
**Stepwise Selection**: Combine forward and backward selection, considering both adding and removing variables.   
Note that the above methods also apply to correlation values. Below, we will see how to use backward elimination.

From the model summary table for the full model with all independent variables above, we can remove variables whose p-value is less than 0.05. This include:  
**ZN** (proportion of residential land zoned for lots over 25,000 sq.ft)  
**INDUS** (proportion of non-retail business acres per town)  
**AGE** (proportion of owner-occupied units built prior to 1940)  
Then, we can build another model without those variables.

```python
Y = df['MEDV']
X = df.drop(['MEDV', 'ZN', 'INDUS', 'AGE'],axis=1)
# Add a constant term to the independent variables matrix for the intercept
X = sm.add_constant(X)

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 

# Fit the multiple linear regression model
model = sm.OLS(Y_train, X_train).fit()
#model = sm.OLS(Y, X).fit()

summary = model.summary()
print(summary)

Y_pred = model.predict(X_test) # Generate predictions

mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r_squared = r2_score(Y_test, Y_pred)

print('\nEvaluation on Testing Set:')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
print(f'R-squared (R): {r_squared:.2f}')
```

From the result we can see by removing three irrelavant independent variables, the new model has an R2 of 0.65, which is close to the R2 of the full model (0.67)
